```python
import pandas as pd
```

```python
sbti = pd.read_excel('companies-taking-action.xlsx')
```

```python
sbti.describe()
```

```python
missing_values = sbti.isna()

missing_values_count = missing_values.sum()

total_missing_count = missing_values.sum().sum()

print("Missing values in each column:")
print(missing_values_count)

print("\nTotal missing values in the DataFrame:", total_missing_count)
```

```python
class_counts = sbti['Sector'].value_counts()
print(class_counts)
```
