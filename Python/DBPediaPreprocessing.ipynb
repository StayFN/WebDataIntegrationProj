```python
import requests

# API endpoint for USD rates
BASE_URL = "https://open.er-api.com/v6/latest/USD"

# Get data from the API
response = requests.get(BASE_URL)
data = response.json()

# Extract exchange rates for your list of currencies
currency_symbols = [  # Based on the previous mapping
    'EUR', 'USD', 'GBP', 'BDT', 'JPY', 'LKR', 'KRW', 'CAD', 'DKK', 'NOK', 'CHF', 'SEK', 
    'HRK', 'INR', 'PHP', 'THB', 'MVR', 'AUD', 'CNY', 'AED', 'RON', 'MXN', 'LVL', 'MYR', 
    'GEL', 'ISK', 'CZK', 'BGN', 'RUB', 'NGN', 'KES', 'ZAR', 'AZN', 'DZD', 'HKD', 'PKR', 
    'GMD', 'NZD', 'RWF', 'SAR', 'DJF', 'ALL', 'SGD', 'HUF', 'IDR', 'UAH', 'TWD', 'OMR', 
    'AMD', 'BWP', 'TRY', 'ARS', 'MAD', 'KZT', 'GHS', 'ETB', 'MWK', 'UGX', 'AFN', 'BRL', 
    'COP', 'MDL', 'QAR', 'ILS', 'KWD', 'JOD', 'BHD', 'PEN', 'CLP', 'MUR', 'MOP', 'JMD', 
    'PGK', 'XOF', 'NPR', 'NAD'
]

# Create an exchange rate table
exchange_rate_table = {currency: data['rates'].get(currency, None) for currency in currency_symbols}

print(exchange_rate_table)
```

```python
import re

def convert_to_usd(value):
    # Check if the value is not a string, and if so, return it as is.
    if not isinstance(value, str):
        return value

    # Try matching the scientific notation first.
    match_sci = re.search(r"([\d+\.Ee\-]+).+datatype/(\w+)>", value)
    # Try matching the regular float notation next.
    match_float = re.search(r"([\d+\.]+).+datatype/(\w+)>", value)

    # Check which match succeeded and extract amount and currency.
    if match_sci:
        amount, currency_name = float(match_sci.group(1)), match_sci.group(2).lower()
    elif match_float:
        amount, currency_name = float(match_float.group(1)), match_float.group(2).lower()
    else:
        return value  # Return the original value if it can't be parsed

    # Convert currency_name to ISO code and then get the exchange rate to USD
    iso_code = currency_map.get(currency_name)
    exchange_rate = exchange_rate_table.get(iso_code, 1)  # Default to 1 if not found

    return amount * exchange_rate
```

```python
df['revenue'] = df['revenue'].apply(convert_to_usd)
df['assets'] = df['assets'].apply(convert_to_usd)
df['netIncome'] = df['netIncome'].apply(convert_to_usd)
df.head()
```

Lets do the Industries Column and convert from DBPedia Links to string values

```python
import numpy as np

def extract_name_from_url(url):
    # If the string doesn't start with 'http://dbpedia.org/resource/', it's probably already extracted.
    if not url.startswith('http://dbpedia.org/resource/'):
        return url

    # Splitting on '/' and taking the last part, then replacing underscores with spaces
    return url.split('/')[-1].replace('_', ' ')

def extract_dbpedia_value(value):
    # Checking if the value is NaN or not a string
    if isinstance(value, float) and np.isnan(value):
        return value
    
    # Handling cases of the format '"value"@en'
    if value.startswith('"') and value.endswith('@en'):
        return value[1:-4]
    
    # Splitting the string on commas
    urls_or_strings = value.split(', ')
    # Extracting names from URLs or keeping the string value, then joining them
    return ', '.join([extract_name_from_url(item) for item in urls_or_strings])
```

```python
df['industries'] = df['industries'].apply(extract_dbpedia_value)
df.head()
```

Lets handle the KeyPeople Column, for some there is an empty string at the start so they start with a ", "  or "* " and "*" is sometimes used as a seperator lets remove that.

```python
df['keyPeople'] = df['keyPeople'].str.lstrip(', ')
df['keyPeople'] = df['keyPeople'].str.lstrip('* ')

df['keyPeople'] = df['keyPeople'].apply(extract_dbpedia_value)
df['keyPeople'] = df['keyPeople'].str.replace('*', ',')
df.head()
```

Now lets extract the year out of the values in the founding year column and replace the values

```python
def extract_year(value):
    if isinstance(value, float) and np.isnan(value):
        return value
    match = re.search(r'(\d{4})', value)
    return match.group(1) if match else value

df['foundingYear'] = df['foundingYear'].apply(extract_year)
df.head()
```
