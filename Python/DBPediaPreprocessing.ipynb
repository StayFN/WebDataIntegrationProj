```python
import pandas as pd
```

```python
tables_1 = pd.read_html(r"../Data/DBPedia_Companies_0-10000.html")
tables_2 = pd.read_html(r"../Data/DBPedia_Companies_10000-MAX.html")
```

```python
df_1 = tables_1[0]
df_2 = tables_2[0]
print(f"df_1 has {len(df_1)} entities")
print(f"df_2 has {len(df_2)} entities")
```

```python
concatenated_df = pd.concat([df_1, df_2], ignore_index=True)
```

```python
concatenated_df.index = ["DBPedia_" + str(i) for i in range(len(concatenated_df))]
concatenated_df
```

```python
#Export as csv with index
concatenated_df.to_csv(r"../Data/DBPedia_Companies_UniqueIndex_NoPreprocessing.csv", index=True)
```
