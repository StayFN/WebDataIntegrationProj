```python
import pandas as pd
#Load DBPedia_Companies_Preprocessed.csv with pandas
df = pd.read_csv(r'../Data/DBPedia_Companies_Preprocessed.csv', sep=',', encoding='utf-8', index_col=0)
df.head()
```

```python
# Convert non-NaN values in 'foundingYear' to strings, and keep NaNs as is
df['foundingYear'] = df['foundingYear'].apply(lambda x: str(int(x)) if pd.notna(x) else x)
```

```python
df.head()
```

```python
cols = ['revenue', 'assets', 'netIncome']

for col in cols:
    # Convert the column to string type
    df[col] = df[col].astype(str)
    
    # Split by '.' and keep only the integer part
    df[col] = df[col].str.split('.').str[0]

df.head()
```

```python
# Replace NaN values with empty string
df = df.fillna("")
df.index.name = "UID"
cols = ['revenue', 'assets', 'netIncome']

df[cols] = df[cols].replace("nan", "")

```

```python
#Save the dataframe to a csv file
df.to_csv(r'../Data/DBPedia_Companies_Preprocessed.csv', index=True, sep=',', encoding='utf-8', na_rep="")
```
