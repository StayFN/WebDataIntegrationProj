```python
import pandas as pd
#Load DBPedia_Companies_Preprocessed.csv with pandas
df = pd.read_csv(r'../Data/DBPedia_Companies_Preprocessed.csv', sep=',', encoding='utf-8', index_col=0)
df.head()
```

```python
# Convert non-NaN values in 'foundingYear' to strings, and keep NaNs as is
df['foundingYear'] = df['foundingYear'].apply(lambda x: str(int(x)) if pd.notna(x) else x)

```

```python
df.head()
```

```python
# Convert the float strings to integer strings for each column
cols = ['revenue', 'assets', 'netIncome']

for col in cols:
    df[col] = df[col].str.split('.').str[0]

df.head()
```

```python
#Save the dataframe to a csv file
df.to_csv(r'../Data/DBPedia_Companies_Preprocessed.csv', sep=',', encoding='utf-8')
```
